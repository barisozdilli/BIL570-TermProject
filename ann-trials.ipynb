```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import matplotlib.pyplot as plt
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device = torch.device('cpu')
device
```

```python
selected_features = ['reanalysis_specific_humidity_g_per_kg',
                     'reanalysis_dew_point_temp_k',
                     'station_avg_temp_c',
                     'station_min_temp_c'
                    ]
```

```python
def prepare_data(data_path, city_abb, train_ratio=0.8, batch_size=2000):

    train_features = pd.read_csv(data_path + '/dengue_features_train.csv', index_col=[0,1,2])
    test_features = pd.read_csv(data_path + '/dengue_features_test.csv', index_col=[0,1,2])
    train_labels = pd.read_csv(data_path + '/dengue_labels_train.csv', index_col=[0,1,2])

    city_train_count = train_features.loc[city_abb].shape[0]
    city_data = pd.concat([train_features.loc[city_abb], test_features.loc[city_abb]], axis=0)
    city_data = city_data[selected_features]
    city_data = city_data.fillna(method = 'pad')
    city_labels = train_labels.loc[city_abb]
    city_val_limit = int(city_train_count * train_ratio)
    city_train_features = city_data[:city_val_limit]
    city_val_features   = city_data[city_val_limit:city_train_count]
    city_test_features  = city_data[city_train_count:]
    city_train_labels   = city_labels[:city_val_limit]
    city_val_labels     = city_labels[city_val_limit:]
    city_train_data = TensorDataset(torch.tensor(city_train_features.values), torch.tensor(city_train_labels.values).squeeze())
    city_val_data   = TensorDataset(torch.tensor(city_val_features.values), torch.tensor(city_val_labels.values).squeeze())
    city_test_data  = TensorDataset(torch.tensor(city_test_features.values))

    city_train_loader = DataLoader(city_train_data, shuffle=True, batch_size=batch_size)
    city_val_loader   = DataLoader(city_val_data, shuffle=True, batch_size=batch_size)
    city_test_loader  = DataLoader(city_test_data, shuffle=True, batch_size=batch_size)
    
    return city_train_loader, city_val_loader, city_test_loader
```

```python
def prepare_data_w_shift(data_path, city_abb, train_ratio=0.8):

    train_features = pd.read_csv(data_path + '/dengue_features_train.csv', index_col=[0,1,2])
    test_features = pd.read_csv(data_path + '/dengue_features_test.csv', index_col=[0,1,2])
    train_labels = pd.read_csv(data_path + '/dengue_labels_train.csv', index_col=[0,1,2])

    city_train_count = train_features.loc[city_abb].shape[0]
    city_data = pd.concat([train_features.loc[city_abb], test_features.loc[city_abb]], axis=0)
    city_data = city_data[selected_features]
    city_data = city_data.fillna(method = 'pad')

    city_data['previous_week'] = train_labels.loc[city_abb]
    city_data['previous_week'] = city_data['previous_week'].shift(+1)
    city_data['previous_week'].fillna(method='backfill', inplace=True)

    city_labels = train_labels.loc[city_abb]
    city_val_limit = int(city_train_count * train_ratio)
    city_train_features = city_data[:city_val_limit]
    city_val_features   = city_data[city_val_limit:city_train_count]
    city_test_features  = city_data[city_train_count:]
    city_train_labels   = city_labels[:city_val_limit]
    city_val_labels     = city_labels[city_val_limit:]
    city_train_data = TensorDataset(torch.tensor(city_train_features.values), torch.tensor(city_train_labels.values).squeeze())
    city_val_data   = TensorDataset(torch.tensor(city_val_features.values), torch.tensor(city_val_labels.values).squeeze())
    city_test_data  = TensorDataset(torch.tensor(city_test_features.values))
    
    return [torch.tensor(city_train_features.values).to(torch.float32),
            torch.tensor(city_val_features.values).to(torch.float32),
            torch.tensor(city_test_features.values).to(torch.float32),
            torch.tensor(city_train_labels.values).to(torch.float32).squeeze(),
            torch.tensor(city_val_labels.values).to(torch.float32).squeeze()
           ]
```

```python
class Network(nn.Module):
    def __init__(self, num_features):
        
        super(Network, self).__init__()
        
        self.fc1 = nn.Linear(in_features=num_features, out_features=50)
        self.drop1 = nn.Dropout(p=0.4)
        
        self.fc2 = nn.Linear(in_features=50, out_features=10)
        self.drop2 = nn.Dropout(p=0.2)

        self.out = nn.Linear(in_features=10, out_features=1)
        
    def forward(self,t):
        
        t = self.fc1(t)
        t = self.drop1(t)
        t = F.relu(t)
        
        t = self.fc2(t)
        t = self.drop2(t)
        t = F.relu(t)
        
        t = self.out(t)
        
        return t
```

```python
test_outputs_shift = torch.tensor([])
cities = ['iq']
running_losses_val = []
running_losses_tra = []

for city in cities:
    trn_fea, val_fea, tst_fea, trn_lab, val_lab = prepare_data_w_shift('./data', city, train_ratio=0.95)
    model = Network(5)
    lr = 0.0001
    criterion = nn.L1Loss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    epoch = 100
    model.train()
    print_period = 1

    for i in range(epoch):
        model.train()
        running_loss_tra = 0.0
        for j in range(trn_fea.shape[0]):
            inputs = trn_fea[j,:].unsqueeze(0)
            labels  = trn_lab[j]
            model.zero_grad()
            output = model(inputs)
            loss = criterion(output.squeeze(), labels)
            running_loss_tra += loss.item()
            loss.backward()
            optimizer.step()
        running_loss_tra /= trn_fea.shape[0]
        running_losses_tra.append(running_loss_tra)
        if (i % print_period == (print_period-1)):
            model.eval()
            running_loss_val = 0.0
            for j in range(val_fea.shape[0]):
                inputs = val_fea[j,:].unsqueeze(0)
                labels = val_lab[j]
                output = model(inputs)
                loss = criterion(output.squeeze(), labels)
                running_loss_val += loss.item()
            running_loss_val /= val_fea.shape[0]
            running_losses_val.append(running_loss_val)
#             print(f'city: {city}\t epoch: {i+1}\t loss: {running_loss_val}')
            model.train()
    
    model.eval()
    output = tst_fea[0][4]
    for i in range(tst_fea.shape[0]):
        tst_fea[i,4] = output
        inputs = tst_fea[i,:].unsqueeze(0)
        output = model(inputs)
        test_outputs_shift = torch.cat([test_outputs_shift, output])
```

```python
plt.plot(running_losses_tra, label='train set')
plt.plot(running_losses_val, label='valid set')
plt.xlabel('epoch')
plt.ylabel('MAE Loss')
plt.legend()
plt.savefig('ANN lr=1e-4 5-10-1 IQ')
plt.show()
```

```python
test_outputs = torch.tensor([])
cities = ['iq']
running_losses_val = []
running_losses_tra = []

for city in cities:
    train_loader, val_loader, test_loader = prepare_data('./data', city, train_ratio=0.95)

    model = Network(len(selected_features))
    lr = 0.0001
    criterion = nn.L1Loss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    epoch = int(2e3)
    model.train()
    print_period = 1

    for i in range(epoch):
        running_loss_tra = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device).to(torch.float32), labels.to(device)
            model.zero_grad()
            output = model(inputs)
            loss = criterion(output.squeeze(), labels.to(torch.float32))
            running_losses_tra.append(loss.item())
            loss.backward()
            optimizer.step()
        if (i % print_period == (print_period-1)):
            model.eval()
            for inputs, labels in val_loader:
                inputs, labels = inputs.to(device).to(torch.float32), labels.to(device)
                output = model(inputs)
                loss = criterion(output.squeeze(), labels.to(torch.float32))
                running_losses_val.append(loss.item())
#                 print(f'city: {city}, epoch: {i+1}, loss: {loss.item()}')
            model.train()

    model.eval()
    for inputs in test_loader:
        inputs = inputs[0].to(device).to(torch.float32)
        output = model(inputs)
        test_outputs = torch.cat([test_outputs, output])
```

```python
plt.plot(test_outputs_shift.squeeze().detach().numpy(), label='4 features model')
plt.plot(test_outputs.squeeze().detach().numpy(), label='5 features model')
plt.xlabel('test sample')
plt.ylabel('prediction')
plt.legend()
plt.savefig('Prediction 4,5-50-10-1 IQ')
plt.show()
```
